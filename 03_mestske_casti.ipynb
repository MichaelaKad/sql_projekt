### Názvy městských částí
Join s exportem číselníku ČSÚ z Tableau od mentorky Elišky

```sql
CREATE OR REPLACE TABLE "vycistena_tabulka_oprava_2" AS
SELECT * EXCLUDE "Id"
FROM "vycistena_tabulka_oprava"
LEFT JOIN "IDs" ON "vycistena_tabulka_oprava"."id" = "IDs"."Id";
```
